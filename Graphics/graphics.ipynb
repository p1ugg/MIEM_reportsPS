### Графический отчет «кластеризованная столбчатая диаграмма» для пары «качественный атрибут —качественный атрибут»

````python
data = {
    'Country': BOOKS['Country'].tolist(),
    'Genre': BOOKS['Genre'].tolist()
}

df = pd.DataFrame(data)
# Исключение данных по стране "США"
df = df[df['Country'] != 'USA']
count_data = df.groupby(['Country', 'Genre']).size().unstack()

# Создание кластеризованной столбчатой диаграммы
count_data.plot(kind='bar', stacked=False)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Clustered Bar Chart for Country and Genre')
plt.legend(title='Genre')
plt.show()
````

````python
data = {
    'Genre': BOOKS['Genre'].tolist(),
    'Country': BOOKS['Country'].tolist()
}

df = pd.DataFrame(data)
# Исключение данных по стране "США"
df = df[df['Country'] != 'USA']

count_data = df.groupby(['Genre', 'Country']).size().unstack()

# Создание кластеризованной столбчатой диаграммы
count_data.plot(kind='bar', stacked=False)
plt.xlabel('Genre', fontsize=2)
plt.ylabel('Country', fontsize=2)
plt.title('Clustered Bar Chart for Genre and Country')
plt.legend(title='Country')
plt.legend(fontsize=7)
plt.show()
````

###Графический отчет «категоризированная гистограмма» для пары «количественный атрибут—качественный атрибут»

````python
data = {
    'User Rating': BOOKS['User Rating'].tolist(),
    'Genre': BOOKS['Genre'].tolist()
}

df = pd.DataFrame(data)
# Группировка данных по категориальному атрибуту и построение гистограммы
plt.figure(figsize=(8, 6))
for category in df['Genre'].unique():
    subset = df[df['Genre'] == category]
    plt.hist(subset['User Rating'], alpha=0.5, label=category)

plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.title('Categorized Histogram for User Rating by Genre')
plt.legend(title='Genre')
plt.show()
````

##Текстовый отчет №1

````python
def get_most_expensive_books(df: pd.DataFrame, Count: int) -> pd.DataFrame:
  """
    Функция генерации отчетов о 20 самых дорогих книгах.

    Parameters
    ----------
    df : pd.DataFrame
        Исходная таблица с данными.

    Count : int
        Количество возвращаемых книг

    Returns
    -------
    pd.DataFrame
        Отчет о 20 самых дорогих книгах.
    """
  data = {
      "Name": BOOKS['Name'].tolist(),
      "Author": BOOKS['Author'].tolist(),
      "Price": BOOKS['Price'].tolist(),
  }
  df = pd.DataFrame(data)

  # Выберем 20 самых дорогих книг
  top_20_expensive_books = df.nlargest(Count, 'Price')

  return top_20_expensive_books

get_most_expensive_books(BOOKS, 20)
````

##Текстовый отчет №2

````python
def get_most_discussed_authors(df: pd.DataFrame, Count: int) -> pd.DataFrame:
  """
    Функция генерации отчетов о 20 самых обсуждаемых авторов.

    Parameters
    ----------
    df : pd.DataFrame
        Исходная таблица с данными.

    Count : int
        Количество авторов

    Returns
    -------
    pd.DataFrame
        Отчет о 20 самых обсуждаемых авторов.
    """
  # Группируем данные по авторам и считаем среднее количество отзывов
  author_reviews = df.groupby('Author')['Reviews'].mean().reset_index()

  # Выбираем топ авторов по среднему количеству отзывов
  top_authors = author_reviews.nlargest(Count, 'Reviews')

  return top_authors

get_most_discussed_authors(BOOKS, 15)
````

##Текстовый отчет №3

````python
# Функция для расчета среднего рейтинга на страну
def get_average_rating_by_country(df: pd.DataFrame) -> pd.DataFrame:
    """
    Функция распределение среднего рейтинга в зависимости от страны.

    Parameters
    ----------
    df : pd.DataFrame
        Исходная таблица с данными.

    Returns
    -------
    pd.DataFrame
        Отчет о распределении среднего рейтинга в зависимости от страны
    """

    country_rating = df.groupby('Country')['User Rating'].mean().reset_index()
    return country_rating

get_average_rating_by_country(BOOKS)
````

##Текстовый отчет №4

````python
def get_average_price_per_year(df: pd.DataFrame, start_year: int, end_year: int) -> pd.DataFrame:
    """
    Функция распределения средней стоимости книги в период с 2009 по 2019 года.

    Parameters
    ----------
    df : pd.DataFrame
      Исходная таблица с данными.

    Returns
    -------
    pd.DataFrame
      Отчет о распределении средней стоимости книги в период с 2009 по 2019 года
    """

    # Фильтруем данные по указанному периоду
    df_filtered = df[(df['Year'] >= start_year) & (df['Year'] <= end_year)]

    # Преобразуем столбец "Price" в числовой формат
    df_filtered['Price'] = pd.to_numeric(df_filtered['Price'], errors='coerce')

    # Группируем данные по годам и считаем среднюю стоимость
    year_price = df_filtered.groupby('Year')['Price'].mean().reset_index()

    return year_price

get_average_price_per_year(BOOKS, 2009, 2019)
````
